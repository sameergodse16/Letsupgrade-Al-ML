# Bank Loan Data 

```python
import pandas as pd
data=pd.read_excel("Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
```

```python
data.head()
```

```python
data.drop(['ID', 'ZIPCode'], axis=1)
```

```python
data.isnull()
```

```python
# Defining the Axises

y=data.PersonalLoan
x=data[['Age','Experience','Income','Family','CCAvg','Education','Mortgage','SecuritiesAccount','CDAccount','Online', 'CreditCard']]
```

```python
data.info()
```

```python
data.describe()
```

```python
import statsmodels.api as sm
x1=sm.add_constant(x)
logistic=sm.Logit(y,x1)
```

```python
# Finding out the
result=logistic.fit()
```

```python
result.summary()
```

# Conclusion:-  

# Most significant variables: If a person wants to get a bank loan these below six variables are more importnace:- Income,Family,Education,CD account, Online, Credit Card as the P value is lesser than 0.05%.

#  Moderate significant variables:- CC Avg and Securities Account are less significant then the above 6 but more significant than Age,Experince and Mortgage.

# Less or No  significant variables: All 3 that is Age,Experince and Mortgage can be ignored as the are greater than P value 0.05%
