```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=1,max_features=2,oob_score=True)
```

```python
features=['Age','Experience','Income','Family','CCAvg','Education','Mortgage','SecuritiesAccount','CDAccount','Online', 'CreditCard']
```

```python
rf_model.fit(X=data[features],y=data[["PersonalLoan"]])
```

```python
print("OOB Accuracy")
print(rf_model.oob_score);
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(features,imp);
```
