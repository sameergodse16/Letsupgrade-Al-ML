```python
# Answer 2
#1. len -> returns lenght of string
print("String function 1:")
LenghtOfSentence = " Calculate length of this sentence"
print("Lenght of above sentence is", len(LenghtOfSentence))
LenghtOfWord = 'Hello'
print ("Lenght of above word is:",len(LenghtOfWord))
print("\n")

#2. format()-> format the specidfied value and inserts it inside strings  placeholder, placeholder is defined using  curly brackets {}
print("String function 2:")
txt = "My name is {Myname}, I like {Coding}".format(Myname = "Sameer", Coding = "Python")
print(txt)
print("\n")

#3. find() -> Finds the first index of a substring in given string
print("String function 3:")
Findstr = "Find the word Hello in this sentence"
Result = Findstr.find("Hello")
print("Find the word 'Hello':", Result)
print("\n")

#4. count() -> it tells us number of times the specific value has occured
print("String function 4:")
CountValue = "This is a string isnt it?"
print(" The word 'is' has appeared ", CountValue.count('is') , "times")
print("\n")

#5. upper() -> Converters String to uppercase
print("String function 5:")
str = "String was lowercase, this will convert to upper case"
Result = str.upper()
print("String converted to UpperCase", Result )
print("\n")

#6. lower() -> Converters String to lowercase
print("String function 6:")
str = "String was uppercase, THIS WILL CONVERT TO LOWER CASE"
Result = str.lower()
print("String converted to LowerCase", Result )
print("\n")

#7. capitalize() - > Coverts the first character upper case
print("String function 7:")
str = "hello, My name is Sameer!"
Result = str.capitalize()
print("Convert First characters, result is:",Result)
print("\n")

#8. replace() - >  is used to create a new string by replacing a specific part from that string
print("String function 8:")
str = "I love Perl"
Result = str.replace("Perl", "Python")
print("New String is:", Result)
print("\n")

#9.   islower() - > this returns true is all the characters are in lower case else it returns false
print("String function 9:")
str1 = "this all is lowercase"
Result = str1.islower()
print ("This is TRUE:", Result)
str2 = "this all iS NOT LowerCase"
Result = str2.islower()
print ("This is False:", Result)
print("\n")

#10. isupper() - > this returns true is all the characters are in lower case else it returns false
print("String function 10:")
str1 = "THIS IS ALL UPPERCASE"
Result = str1.isupper()
print ("This is TRUE:", Result)
str2 = "This all is NOT UpperCAse"
Result = str2.isupper()
print ("This is False:", Result)
print("\n")

#11. isdigit() ->  Returns True if all charactors in string are decimal charactors else it returns false
print("String function 11:")
int1 = '100'
print ("This will return TRUE",int1.isdigit())

int2 = '10.5'
print ("This will return False as dot is not a digit", int2.isdigit())

print("\n")

#12. startswith() -> function returns True if the string ends with the given prefix, otherwise it returns False.
print("String function 12:")
str = "Smita, welcome to India."
print(str.startswith("wel",7,20))
print(str.startswith("wel",7,9))
print(str.startswith("wel",7,10))
print(str.startswith("wel",6,9))
print("\n")

#13.  returns a new string after removing any leading and trailing whitespaces including tabs (\t).
print("String function 13:")
str = "     Space  ."
Result = str.strip()
print("Result after strip: ", Result)
print("\n")

#14.  swapcase() ->  this converts uppercase to lower and lowercase to upper
print("String function 14:")
str = "This is upper AND LOWER case string"
Result = str.swapcase()
print ("Converted string is:", Result)
print("\n")

#15. split() ->is used to split a string into the list of strings based on a delimiter.
print("String function 15:")
str = "I*Love*SugarCane*Juice"
Result = str.split("*")
print ("Splitted String is:", Result)
print("\n")
```

```python
# Answer 3 for polindrome
string1=input("Enter the word:")
print(string1)
rev_string=string1[::-1]
if rev_string==string1:
    print("polindrome")
else:
    string2=input("Enter the word:")
    str1=sorted(string1)
    str2=sorted(string2)
    if str1==str2:
        print("Anagrams")

    else:
            print("None")
```
