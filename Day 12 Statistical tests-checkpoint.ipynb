```python
import pandas as pd
dataset=pd.read_excel("general_data1.xlsx",sheet_name=0)
dataset.head()
```

```python
dataset["Attrition"]=dataset["Attrition"].map({"Yes":1,"No":0})
dataset.head()
```

```python
# Wilcoxon test
# Ho = There is no significant difference NumCompaniesWorked and Attrition
# Ha = There is  significant difference NumCompaniesWorked and Attrition
from scipy.stats import wilcoxon
d1=dataset.NumCompaniesWorked
d2=dataset.Attrition
stat, p=wilcoxon(d1,d2)
print(stat,p)
# As p-value  In this case is zero we are reject the null hypothesis because the significance is . 000,
# which is substantially less
```

```python
# Friedman  chi square Test
# Ho = There is no significant difference NumCompaniesWorked and Attrition and Years Since Last Promotion
# Ha = There is  significant difference NumCompaniesWorked and Attrition and Years Since Last Promotion
from scipy.stats import friedmanchisquare
d3=dataset.YearsSinceLastPromotion
stat,p=friedmanchisquare(d1,d2,d3)
print(stat,p)
# As p-value  In this case is zero we are reject the null hypothesis because the significance is . 000,
# which is substantially less
```

```python
# Mann whitney Test
# Ho = There is no significant difference NumCompaniesWorked and Attrition and Years Since Last Promotion
# Ha = There is  significant difference NumCompaniesWorked and Attrition and Years Since Last Promotion

from scipy.stats import mannwhitneyu
d3=dataset.Attrition
d4=dataset.JobLevel
stat,p=mannwhitneyu(d3,d4)
print(stat,p)
# As p-value  In this case is zero we are reject the null hypothesis because the significance is . 000,
# which is substantially less
```
