```python
data.fillna('1',inplace=True)
```

```python
data.isnull()
```

```python
data.head()
```

```python
# Import label encoder 
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
data["Attrition"]=encode.fit_transform(data["Attrition"])
data["BusinessTravel"]=encode.fit_transform(data["BusinessTravel"])
data["Department"]=encode.fit_transform(data["Department"])
data["EducationField"]=encode.fit_transform(data["EducationField"])
data["Gender"]=encode.fit_transform(data["Gender"])
data["JobRole"]=encode.fit_transform(data["JobRole"])
data["MaritalStatus"]=encode.fit_transform(data["MaritalStatus"])
```

```python
data.head()
```

```python
# Defining the Axises

y=data.Attrition
x=data[['Age','Department','JobRole','MaritalStatus','Education','MonthlyIncome','Gender','PercentSalaryHike','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
data.info()
```

```python
import statsmodels.api as sm
x1=sm.add_constant(x)
logistic=sm.Logit(y,x1)
```

```python
result=logistic.fit()
```

```python
result.summary()
```

# Conclusion:-  

# Most significant variables: The below factors Age,Marital Status,Training Times Last Year,Years Since Last Promotion and Years With Curr Manager are more significant for employee attrition as these are lesser than the P value 0.05%

#  Moderate significant variables:   MonthlyIncome,Department,Job Role,Years At Company have moderate signifiance.

# Less or No  significant variables: Education and Gender can be ignored as the are greater than P value 0.05%
